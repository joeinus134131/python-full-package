Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.stats import linregress
```

```python
x = np.random.randint(3,20,size=20)
y = np.random.randint(20,30,size=20)
```

```python
dataPlot = pd.DataFrame(zip(x,y),columns=['x','y'])
```

```python
dataPlot.plot(kind='line')
```

### Menghitung nilai slope dan intercept dari linear regression

```python
slope,intercept,r,p,se=linregress(x,y)
```

```python
slope
```

```python
intercept
```

```python
r
```

```python
p
```

```python
se
```

```python
def manual_regression(x):
    return slope*x+intercept
```

```python
manual_regression(1)
```

```python
list(map(manual_regression,x))
```

```python
plt.plot(x,list(map(manual_regression,x)))
plt.plot(x,y,'o')
```

### Processing Data From a File

```python
rumah = pd.read_excel('DATASETS/rumah.xlsx')
```

```python
rumah
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

rumah['Location'] = le.fit_transform(rumah['Location'])

rumah
```

```python
rumah.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = rumah[['Area','Bedrooms','Location']]
y = rumah['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
X_test
```

```python
predict_value = model.predict(X_test)
predict_value
```

```python

new_data = pd.DataFrame([[300, 6, 1]], columns=['Area', 'Bedrooms', 'Location'])
```

```python
predict_price = model.predict(new_data)
predict_price
```

```python
date = pd.read_csv('DATASETS/date.csv', header=None, names=['bulan','tabungan'])
```

```python
date
```

```python
date['bulan_inum'] = date['bulan'].index+1
```

```python
date
```

```python
date.info()
```

```python
X = date[['bulan_inum']].iloc[:53]
y = date[['tabungan']].iloc[:53]
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.predict(date[['bulan_inum']].loc[54:])
```

```python
plt.plot(date[['bulan_inum']],lr.predict(date[['bulan_inum']]))
plt.plot(date[['tabungan']],'o')
plt.title('Model Linear Regression')
plt.xlabel('Bulan')
plt.ylabel('Tabungan')
plt.show()
```

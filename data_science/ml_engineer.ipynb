### Credit history

jika credit_history nan dan loan_amount nan maka kredit history 0
jika credit_history nan maka kredit history 1

```python
def fill_credit_history(row):
    """Fill the credit history of a customer with a random number of transactions."""
    if str(row['Credit_History'])=='nan' and str(row['LoanAmount'])=='nan':
        return 0
    elif str(row['Credit_History'])=='nan' and str(row['LoanAmount'])!='nan':
        return 1
    else:
        return row['Credit_History']
```

```python
df['Credit_History'].unique()
```

```python
df['Credit_History']=df.apply(fill_credit_history,axis=1)
```

```python
df.info()
```

## Gender
jika applicant income > co applicant income maka dia male
jika applicant income < co applicant income maka dia female

```python
df['Gender'].unique()
```

```python
def fill_gender(row):
    """Fill the credit history of a customer with a random number of transactions."""
    if str(row['Gender'])=='nan' and str(row['LoanAmount'])=='nan':
        return 0
    elif str(row['Credit_History'])=='nan' and str(row['LoanAmount'])!='nan':
        return 1
    else:
        return row['Credit_History']
```

```python
df['Gender'] = df.apply(
    lambda x:'Male' if (x['ApplicantIncome']>x['CoapplicantIncome']) and (str(x['Gender']) =='nan') else  \
        ('Female' if (x['ApplicantIncome']<x['CoapplicantIncome']) and str(x['Gender']) =='nan' else x['Gender']),axis=1)
```

```python
df.info()
```

### Married

### Dependents
jika status pernikahan (Married) yes maka dependents nya nilainya 2
jika status pernikahan (Married) no maka dependents nya nilainya 0

```python
df['Dependents'].unique()
```

```python
def change_dependents(row):
    """Change the dependents of a row in the dataframe."""
    if str(row['Married']) == 'Yes' and str(row['Dependents']) == 'nan':
        return 2
    elif str(row['Married'])=='No' and str(row['Dependents']) == 'nan':
        return 0
    else:
        return row['Dependents']
```

```python
df['Dependents'] = df.apply(change_dependents,axis=1)
```

### Loan Amount

```python
df['LoanAmount'] = df['LoanAmount'].fillna(0)
```

```python
max(df['LoanAmount'])
```

```python
def loan_group(row):
    """Return the loan group of the given row."""
    for i,num in enumerate([1,2,3,4,5,6,7]):
        if row <= num*100:
            return i+1
            break
```

```python
loan_group(600)
```

```python
df['LoanAmount']=df['LoanAmount'].apply(loan_group)
```

### Featured Engineering

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.columns
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str)
    df[col] = le.fit_transform(df[col])
    dic[col] = le.classes_
```
